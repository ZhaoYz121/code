```python
#读取npy文件
from numpy import load
true = load('true.npy')
pred = load('pred.npy')
```

```python
x = true
y = pred
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)

idx = z.argsort()
x, y, z = x[idx], y[idx], z[idx]

fig = plt.figure(figsize=(8,6))
plt.scatter(x,y,c=z,s=40,cmap=plt.get_cmap('jet'))
plt.plot(np.arange(-5,30),np.arange(-5,30),c='red',ls='--')
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
plt.xlim(-5,30)
plt.ylim(-5,30)
plt.title('Argo-BLT(m)',fontsize=20)
plt.xlabel('Argo-BLT(m)',fontsize=20)
plt.ylabel('ANN-BLT(m)',fontsize=20)
plt.text(-3,28,'RMSE='+str(round(mean_squared_error(true,pred) ** 0.5,3)),fontsize='15')
plt.text(-3,25,'R$^2$ = '+str(round(r2_score(true,pred),3)),fontsize='15')
cbar = plt.colorbar()
cbar.ax.tick_params(labelsize=17)
```

```python
x = true
y = pred
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)

idx = z.argsort()
x, y, z = x[idx], y[idx], z[idx]
font = {'family':'Times New Roman'}
fig = plt.figure(figsize=(6,4.5))
plt.scatter(x,y,c=z,s=70,cmap=plt.get_cmap('jet'))

# 利用polyfit函数拟合直线
slope, intercept = np.polyfit(x, y, 1)

# 绘制直线
plt.plot(x, slope*x + intercept, c='grey', ls='--', label='Linear Regression')
plt.plot(np.arange(-2,30),np.arange(-2,30),c='red',ls='--')
plt.xticks(fontproperties='Times New Roman')
plt.yticks(fontproperties='Times New Roman')
plt.tick_params(labelsize=15)
plt.xlim(-2,30)
plt.ylim(-2,30)
plt.title('(d) Testing: R$^2$ = 0.85',fontsize=20,fontdict=font)
plt.xlabel('Argo BLT(m)',fontsize=20,fontdict=font)
plt.ylabel('ANN BLT(m)',fontsize=20,fontdict=font)
cbar = plt.colorbar()
cbar.ax.tick_params(labelsize=15)
plt.rcParams['font.family'] = 'Times New Roman'
plt.savefig('图zhengR2.png')
plt.show()
```
